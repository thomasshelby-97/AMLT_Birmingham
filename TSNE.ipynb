### Crating a random sample of 1000 from 70000

```python
np.random.seed(100)
sample = np.random.choice(X.shape[0], 1000)
print (sample)
```

### Creating a new set of 1000

```python
X_1 = X.iloc[sample, :]
X_1
```

```python
y_1 = y[sample]
y_1
```

### Building the t-SNE model

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components = 2, perplexity = 30)
```

### Transforming X_1

```python
X_tsne = tsne.fit_transform(X_1)
X_tsne
```

```python
X_tsne.shape
```

```python
X_tsne[0]
```

### Visualization of the transformed points

```python
plt.scatter(X_tsne[:, 0], X_tsne[:, 1]);
```

```python
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c = y_1.astype(float));
```

### Creating a Dataframe for better visualization

```python
X_df = pd.DataFrame({'X0' : X_tsne[:, 0], 'X1' : X_tsne[:, 1], 'Label' : y_1})
X_df
```

### Plotting the DataFrame

```python
plt.figure(figsize = (15, 12))
sns.lmplot(data = X_df, x = 'X0', y = 'X1', hue = 'Label');
```

```python
plt.figure(figsize = (15, 12))
sns.lmplot(data = X_df, x = 'X0', y = 'X1', hue = 'Label', fit_reg = False);
```
